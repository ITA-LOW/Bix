```python
import pandas as pd
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/ITA-LOW/teste_Bix/main/Base%20de%20dados/Base%20de%20Dados%20-%20Question%C3%A1rio.xlsx%20-%20Sheet1.csv'
df = pd.read_csv(url)
```

```python
df.drop(columns=['Unnamed: 0',	'Unnamed: 1'], inplace=True)
```

```python
df.drop([0,1,2], inplace=True)
```

```python
df.columns=['ID',	'ProductID',	'ClientID',	'Discount',	'UnitPrice',	'Quantity',	'StoreID',	'Date']
```

```python
df.drop([3], inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df.Date,dayfirst = True)
```

```python
df.dtypes
```

```python
df.groupby(['StoreID'], dropna=False).size().sort_values(ascending=False)
```

```python
df['Discount'] = df['Discount'].str.replace(',','.').astype(float)
```

```python
df['UnitPrice'] = df['UnitPrice'].str.replace(',','.').astype(float)
```
